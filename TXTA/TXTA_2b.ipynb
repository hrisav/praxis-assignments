```python
from selenium.webdriver import Chrome
import time
```

```python
from bs4 import BeautifulSoup
```

```python
PATH = 'C:\Program Files (x86)\chromedriver.exe'

driver = Chrome(executable_path=PATH)

url = 'https://housing.com/in/buy/searches/M2P5rq7v7nkml5mhwu9'
driver.get(url)

time.sleep(3)

i=0

while(i<=8):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(3)
    i=i+1
    
time.sleep(3)
containers = [x.get_attribute('innerHTML') for x in driver.find_elements_by_xpath("//div[@class='css-v64o2m']")]
```

```python
type(containers)
```

```python
len(containers)
```

```python
containers[0]
```

```python
pprice = []
pbhk = []
plocation = []

for container in containers:
    doc = BeautifulSoup(container)
    try:
        pprice.append(doc.find('div', {'class':'css-18rodr0'}).text)
    except:
        pprice.append('Price Not Available')
    try:
        pbhk.append(doc.find('div', {'class':'css-ybv4ci'}).text.split('BHK')[0].strip())
    except:
        pbhk.append('BHK Not Available')
    try:
        plocation.append(doc.find('a', {'class':'css-16drx2b'}).text.strip())
    except:
        plocation.append('Location Not Available')
```

```python
import pandas as pd

details = list(zip(pprice, pbhk, plocation))

df = pd.DataFrame(details, columns = ['Price', 'BHK', 'Location'])
df.head()
```
