### Create a dataframe which consists of name of each sales representative along with their sales conversion rate in percentage (based on number of sales). Name this dataframe as "conversion_df".

```python
df_merged.head()
```

```python
conversion_df = pd.crosstab(index=df_merged['sales_rep'], columns=df_merged['sale_converted'])
conversion_df.head()
```

```python
conversion_df.reset_index(inplace=True)
```

```python
conversion_df = conversion_df.rename({0:'not_conv',1:'conv'}, axis=1)
conversion_df.head()
```

```python
conversion_df['conversion_rate'] = round((conversion_df['conv']*100/(conversion_df['not_conv']+conversion_df['conv'])), 2)
conversion_df.head()
```

### With this new dataframe (the one with conversion rates), create three categories for the sales representative based on conversion_rate described as below:


> a) <10% - "low"

> b) >=10% to <20% - "medium"

> c) >=20% - "high"

```python
bins = [0,10,20,np.inf]
labels = ['low','medium','high']

conversion_df['conv_cat'] = pd.cut(conversion_df['conversion_rate'],bins=bins, labels=labels, include_lowest=True, right=False)
```
