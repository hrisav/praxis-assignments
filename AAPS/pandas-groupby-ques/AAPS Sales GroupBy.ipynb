## Question - 4 Using group-by, create a dataframe with the names of top five sales representative who have converted the most number of sales?

```python
df.groupby(['sales_rep']).agg({'sale_converted':'sum'}).reset_index().sort_values(by=['sale_converted'],ascending=False).head(5)
```

## Question - 6 Using group-by, create a dataframe with top 10 sales representative, describing the amount of sales that they were not able to convert? (Descending Order)

```python
df.head(5)
```

```python
df_2 = df.groupby(['sales_rep']).agg({'order_value':'sum','sales_value':'sum'}).reset_index()
```

```python
df_2['missed_sales'] = df_2['order_value'] - df_2['sales_value']
```

```python
df_2.sort_values(by='missed_sales', ascending=False).head(10)
```

```python
df_2["percentage_sale"] = round((df_2['missed_sales']/df_2['order_value'])*100,2)
df_2.sort_values(by='missed_sales', ascending=False).head(10)
```

## Question - 7 Using group-by, create a dataframe, list top ten companies which has created most sales and also mention the corresponding order values placed with them.

```python
df.groupby(['company_name']).agg({'order_value':'sum',
                                 'sales_value':'sum'}).sort_values(by='sales_value', ascending=False).reset_index().head()
```

## Question - 8 Based on the sales value, list out top 5 sales representative and their top 5 companies. Hint: You need a UDF to sort the sales value for each group. OR Use a lambda function to do the same.

```python
df.head(5)
```

```python
df_3 = df.groupby(['sales_rep','company_name']).agg({'sales_value':'sum'}).reset_index()
```

```python
df_3.head()
```

# With UDF

```python
def sort_df(df):
    df = df.copy()
    df = df.sort_values(by=['sales_value'], ascending=False)
    df = df.head(5)
    return df
```

```python
top_5 = df_3.groupby(['sales_rep']).apply(sort_df).reset_index(drop=True)
top_5
```

```python
def sort_df(df: pd.DataFrame):
  df = df.copy()
  df = df.sort_values(by=['sales_value'], ascending=False)
  df = df.head(5)
  return df

top_5 = df_3.groupby(['sales_rep']).apply(sort_df).reset_index(drop=True)
```
